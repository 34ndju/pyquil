```python
import matplotlib.pyplot as plt
import pyquil.forest as forest
```

```python
data = forest.rabi("the_chip", 0.0, 0.75, step=0.026, the_time=200.0)
```

```python
data = data['result']
powers = [dd[0] for dd in data]
amps = [dd[1] for dd in data]
```

```python
plt.title("Power Rabi")
plt.xlabel("Pulse Amplitude (v)")
plt.ylabel("Signal Amplitude (V)")
plt.plot(powers, amps)
plt.show()
```
